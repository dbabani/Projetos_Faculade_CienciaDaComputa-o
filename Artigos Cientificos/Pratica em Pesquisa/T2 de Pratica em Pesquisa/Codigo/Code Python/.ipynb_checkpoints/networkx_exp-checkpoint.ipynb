```python
import matplotlib.pyplot as plt
import networkx as nx
import networkit as nk

colors = ['blue', 'green', 'red', 'yellow']
```

```python
def str_list_to_int(str_list):
  return [int(item) for item in str_list]

def read_edges_from_file(filename):
  with open(filename, "r") as f:
    lines = f.readlines()
    edges = [str_list_to_int(line.split()) for line in lines]
  return edges 

def get_nk_communities():
  nk_graph = nk.graph.Graph()
  nx_graph = nx.Graph()

  for edge in read_edges_from_file('../road-euroroad/road-euroroad.edges'): 
    nk_graph.addEdge(edge[0], edge[1], 0, True)
    nx_graph.add_edge(edge[0], edge[1])

  communities = nk.community.detectCommunities(nk_graph, nk.community.PLP(nk_graph))
  
  nx_communities = {}

  for community in communities.getVector():
    if community not in nx_communities and community != 0:
      nx_communities[community] = {}
    
  for community in nx_communities:
    nx_community = set()

    for member in communities.getMembers(community):
      nx_community.add(member)
    
    nx_communities[community] = nx_community

  return (nx_graph, nx_communities)
```

```python
graph, communities = get_nk_communities()

color_map = []
for node in graph.nodes:
    for (index, community) in enumerate(communities.items()):
        if node in list(community[1]):
            color_map.append(colors[(index + 1) % len(colors)])
            break

nx.draw(graph, node_color=color_map, with_labels=True)
plt.show()
```
